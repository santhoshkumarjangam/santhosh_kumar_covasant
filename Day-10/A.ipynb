```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bikes.csv', index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
# daily attendance for 'Berri1' and 'PierDup'
plt.figure(figsize=(12, 5))
df[['Berri1', 'PierDup']].plot(title="Daily Attendance of Berri1 & PierDup")
plt.xlabel("Date")
plt.ylabel("Number of Bicycles")
plt.show()
```
