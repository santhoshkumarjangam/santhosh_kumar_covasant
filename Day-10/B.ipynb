```python
import pandas as pd
```

```python
df_survival = pd.read_csv('gender_submission.csv')
```

```python
df_survival.head()
```

```python
df_info = pd.read_csv('test.csv')
```

```python
df_info.head()
```

```python
df_main = pd.merge(df_survival, df_info, on="PassengerId")
```

**Analyze survival rate by gender**

```python
gender_survival = df_main.groupby("Sex")["Survived"].mean()
```

```python
print("Survival rate by gender:")
gender_survival
```

**Analyze survival rate by passenger class**

```python
pclass_survival = df_main.groupby("Pclass")["Survived"].mean()
```

```python
print("Survival rate by passenger class:")
pclass_survival
```

**Survival rate by gender and class combined**

```python
gender_pclass_survival = df_main.groupby(["Sex", "Pclass"])["Survived"].mean() * 100
```

```python
print("Survival percentage by gender and passenger class:")
gender_pclass_survival
```
