```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
mean1, sd1 = 0, 1
mean2, sd2 = 2, 1.5
```

```python
data1 = np.random.normal(mean1, sd1 , n)
data2 = np.random.normal(mean2, sd2 , n)
```

```python
data_sum = data1 + data2
```

```python
plt.figure(figsize=(12, 6))

plt.hist(data1, bins=50, density=True, alpha=0.5, label="Gaussian 1 (μ=0, σ=1)")
plt.hist(data2, bins=50, density=True, alpha=0.5, label="Gaussian 2 (μ=2, σ=1.5)")
plt.hist(data_sum, bins=50, density=True, alpha=0.5, label="Sum of two Gaussians", color="red")

plt.show()
```
